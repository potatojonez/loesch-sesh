### 3. Alter function to return the len of list, x list, and y list

```python
def compute_star_shape(s, e = 0.15, a = 3):
    x_list = []
    y_list = []
    i = 0
    while i < 2*math.pi:
        t = (0.5 + e * math.sin(a*i)) * math.cos(i)
        r = (0.5 + e * math.sin(a*i)) * math.sin(i)
        x = t * math.cos(0.25*math.pi) - r * math.sin(0.25*math.pi) + 0.5
        y = t * math.sin(0.25*math.pi) + r * math.cos(0.25*math.pi) + 0.5
        x_list.append(x)
        y_list.append(y)
        i += s
    return len(x_list), x_list, y_list
```

### 4. Plot The graph and calling functions

```python
ds = .01
plt.plot(compute_star_shape(ds)[1], compute_star_shape(ds)[2], color="blue")
plt.xlabel('x')
plt.ylabel('y')
plt.title('e = 0.15, a = 3, # of points = '+str(len(compute_star_shape(ds)[1])))
plt.show()
plt.plot(compute_star_shape(ds, a = 5)[1], compute_star_shape(ds, a = 5)[2], color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.title('e = 0.15, a = 5, # of points = '+str(len(compute_star_shape(ds)[1])))
plt.show()
plt.plot(compute_star_shape(ds, e = 0.3)[1], compute_star_shape(ds, e = 0.3)[2], color="black")
plt.xlabel('x')
plt.ylabel('y')
plt.title('e = 0.3, a = 3, # of points = '+str(len(compute_star_shape(ds)[1])))
plt.show()
plt.plot(compute_star_shape(ds, e = 0.3, a = 5)[1], compute_star_shape(ds, e = 0.3, a = 5)[2], color="green")
plt.xlabel('x')
plt.ylabel('y')
plt.title('e = 0.3, a = 5, # of points = '+str(len(compute_star_shape(ds)[1])))
plt.show()
```

### 5. Let's try something more practical...

I want to determine the projected value of my portfolio over the next x years and draw a graph. <br>
Arguments should include principle, interest rate, # of years, and how frequently the interest is getting compounded. <br><br>
Default interest should be 5%, default compounding should be annually - represent that numerically as 1 (monthly would be 12), term will equal the amount of years * compunding e.g. 30 years compounded annually will equal 30.  Default should be 30 <br><br>
Name your function my_portfolio. <br>
p = principle, i = interest, t = terms, c = compounding <br>
The equation for the value of a 10,000 in the 30th year would be 10000*(1.05) *** 30/1 <br>
Return an x list and y list where x is the term count and y is the value of the portfolio in a given year - up until the end of your selected term length

```python
def my_portfolio(p, i=0.05, t=30, c=1):
    x_list = []
    y_list = []
    for j in range(t+1):
        y = p * (1 + i) ** (j/c)
        x_list.append(j)
        y_list.append(y)
    return(x_list, y_list)
```

### 5.1 Test your equation by finding the value each year with a principle of 10000

```python
my_portfolio(10000)
```
